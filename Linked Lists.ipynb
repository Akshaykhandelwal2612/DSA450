```python
##Write a program to Detect loop in a linked list.

class Solution:
    #Function to check if the linked list has a loop.
    def detectLoop(self, head):
        s=set()
        while head is not None: 
            if head in s:
                return True 
            else:
                s.add(head)
            head=head.next
        return False
    
##or 

class Solution:
    #Function to check if the linked list has a loop.
    def detectLoop(self, head):
        slow=head
        fast=head
        while slow is not None and fast is not None and fast.next is not None: 
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                return True 
        return False
```

```python
##Write a program to Delete loop in a linked list.



class Solution:
    #Function to remove a loop in the linked list.
    def removeLoop(self, head):
        a=set()
        curr=head
        while curr!=None: 
            if curr.next in a:
                curr.next=None
            else:
                a.add(curr)
            curr=curr.next
        return head
```

```python
##Remove Duplicates in a sorted Linked List.

def removeDuplicates(head):
    curr=head
    while curr is not None and curr.next is not None : 
        if curr.data==curr.next.data:
            curr.next=curr.next.next
        else:
            curr=curr.next
    return head
```

```python
## Remove Duplicates in a Un-sorted Linked List.

class Solution:
    #Function to remove duplicates from unsorted linked list.
    def removeDuplicates(self, head):
        d=dict()
        curr=head
        prev=False
        while curr is not None: 
            if curr.data in d.keys():
                prev.next=curr.next
            else: 
                d[curr.data]=0
                prev=curr
            curr=curr.next
        return head
```

```python
### Merge sort for linkedlist
class Solution:
    #Function to sort the given linked list using Merge Sort.
    def getMiddle(self,curr):
        if curr==None:
            return curr
        fast=curr
        slow=curr
        while fast.next is not None and fast.next.next is not None:
            slow=slow.next
            fast=fast.next.next
        return slow
    
    def merger(self,left,right):
        sort=None
        if left==None:
            return right
        if right==None:
            return left
        
        if left.data<right.data:
            sort=left
            sort.next=self.merger(left.next,right)
        else:
            sort=right
            sort.next=self.merger(left,right.next)
        
        return sort
        
        
        
        
    def mergeSort(self, head):
        if head==None or head.next==None:
            return head
            
        middle=self.getMiddle(head)
        fwd=middle.next
        middle.next=None 
       
        left=self.mergeSort(head)
        right=self.mergeSort(fwd)
        
        sort=self.merger(left,right)
        return sort
```

```python
## Middle of LinkedList
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow=head
        fast=head
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
        return slow
```

```python
## Rotate Doubly linked list by N nodes
def rotate(start, N):
    curr=start
    count=0
    while curr is not None :
        count+=1
        curr=curr.next
    if count==N:
        return start
    attach=start
    count=0
    while start.next is not None :
        count+=1
        if count==N+1:
            start.prev.next=None
            start.prev=None
            new_start=start
        start=start.next
    start.next=attach
    attach.prev=start
    return new_start
 
```

```python
## flatten LL 
def merge(a,b):
    if a==None:
        return b
    if b==None:
        return a
    result=None
    
    if a.data<b.data:
        result=a
        result.bottom=merge(a.bottom,b)
    else:
        result=b
        result.bottom=merge(a,b.bottom)
    
    result.next=None
    return result 

def flatten(root):
    if root==None or root.next==None:
        return root 
    root.next=flatten(root.next)
    root=merge(root,root.next)
    return root 
    
```
